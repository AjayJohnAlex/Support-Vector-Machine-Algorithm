# Data Preparation

```python
X = df.drop('spam',axis=1)
y = df['spam'].values.astype(int)
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import scale
```

```python
X = scale(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

# Model Building

```python
from sklearn.svm import SVC
```

```python
model = SVC(C=1, kernel='rbf')
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

# Performance Evaluation

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,mean_absolute_error,mean_squared_error
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(mean_absolute_error(y_test,y_pred))
```

```python
print(mean_squared_error(y_test,y_pred))
```

# HyperParameter Tuning 
We need to tune 3 parameters here :
1. *C*
2. *Type of Kernel*
3. *Gamma*

All 3 would be done by GridSearchCV()

## We need to calculate the value of C, Gamma and decide upon the type of kernel

```python
from sklearn.model_selection import KFold
```

```python
folds = KFold(n_splits=5,shuffle=True,random_state=4)
```

```python
hyper_params = [{ 'gamma' : [1e-2,1e-3,1e-4],
                  'C' : [1,10,100,1000]
}]
```

```python
model = SVC (kernel='rbf')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model_Kernel = GridSearchCV(estimator=model,param_grid=hyper_params,scoring='accuracy',
                            verbose=1,return_train_score=True,cv=folds)
```

```python
model_Kernel.fit(X_train,y_train)
```

```python
model_Kernel.best_params_
```

```python
model_Kernel.best_estimator_
```

```python
model_Kernel.cv_results_
```

```python
kernal_and_Svm_df = pd.DataFrame(model_Kernel.cv_results_)
kernal_and_Svm_df
```

```python
model_Kernel.cv_results_['param_C']
```

```python
# convert C into int value
kernal_and_Svm_df['param_C'] = kernal_and_Svm_df['param_C'].astype('int')
```

```python
model_Kernel
```

```python
# plotting the data 

plt.figure(figsize=(16,6))
# plotting 1/3
plt.subplot(131)
# getting gamma at 0.01 for all values of C (1,10,100,100)
gamma_01 = kernal_and_Svm_df[kernal_and_Svm_df['param_gamma']==0.01]

plt.plot(gamma_01['param_C'],gamma_01['mean_test_score'])
plt.plot(gamma_01['param_C'],gamma_01['mean_train_score'])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title('Gamma = 0.01')
plt.ylim([0.80,1])
plt.legend(['test_accuracy','train_accuracy'],loc='upper left')
plt.xscale('log')


# plotting 2/3
plt.subplot(132)
# getting gamma at 0.001 for all values of C (1,10,100,100)
gamma_001 = kernal_and_Svm_df[kernal_and_Svm_df['param_gamma']==0.001]

plt.plot(gamma_001['param_C'],gamma_001['mean_test_score'])
plt.plot(gamma_001['param_C'],gamma_001['mean_train_score'])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title('Gamma = 0.001')
plt.ylim([0.80,1])
plt.legend(['test_accuracy','train_accuracy'],loc='upper left')
plt.xscale('log')


# plotting 3/3
plt.subplot(133)
# getting all the gamma at 0.0001 for all values of C(1,10,100,100)
gamma_0001 = kernal_and_Svm_df[kernal_and_Svm_df['param_gamma'] == 0.0001]

plt.plot(gamma_0001['param_C'],gamma_0001['mean_test_score'])
plt.plot(gamma_0001['param_C'],gamma_0001['mean_train_score'])
plt.title('Gamma = 0.0001')
plt.xlabel("C")
plt.ylabel('Accuracy')
plt.ylim([0.80,1])
plt.legend(['mean_test_score','mean_train_score'],loc='upper left')
plt.xscale('log')
```

This plot reveals some interesting insights:
- **High values of gamma** lead to **overfitting** (especially at high values of C); note that the training accuracy at gamma=0.01 and C=1000 reaches almost 99% 
- The **training score increases with higher gamma**, though the **test scores are comparable** (at sufficiently high cost, i.e. C > 10)
- The least amount of overfitting (i.e. difference between train and test accuracy) occurs at low gamma, i.e. a quite *simple non-linear model*


```python
model_Kernel.best_score_
```

```python
model_Kernel.best_params_
```

Though sklearn suggests the optimal scores mentioned above (gamma=0.001, C=1000), one could argue that it is better to choose a simpler, more non-linear model with gamma=0.0001. This is because the optimal values mentioned here are calculated based on the average test accuracy (but not considering subjective parameters such as model complexity).

We can achieve comparable average test accuracy (~92.5%) with gamma=0.0001 as well, though we'll have to increase the cost C for that. So to achieve high accuracy, there's a tradeoff between:
- High gamma (i.e. high non-linearity) and average value of C
- Low gamma (i.e. less non-linearity) and high value of C

We argue that the model will be simpler if it has as less non-linearity as possible, so we choose gamma=0.0001 and a high C=1000.
