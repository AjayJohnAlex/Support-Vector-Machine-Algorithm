```python
X = letter_df.drop('letter',axis=1)
y= letter_df['letter']
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import scale
```

```python
X = scale(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

# Model Buliding
Both Linear and Non Linear model are to be build

```python
from sklearn.svm import SVC
```

```python
model_linear  = SVC(C=1,kernel='linear',gamma='auto_deprecated')
model_linear.fit(X_train,y_train)
linear_pred = model_linear.predict(X_test)
```

# hyper parameter tuning 

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import KFold
```

```python
fold = KFold(n_splits=5,random_state=4,shuffle=True)
```

```python
hyper_params = [{ 'C': [1,10,100,1000],
                 'gamma' : [1e-2,1e-3,1e-4]
                
                }]
```

```python
model_nL = SVC(kernel='rbf')
model_non_linear_GS = GridSearchCV(estimator=model_nL,param_grid=hyper_params,cv=fold,scoring='accuracy',
                                   verbose=1,return_train_score=True)
```

```python
model_non_linear_GS.fit(X_train,y_train)
```

```python
GCV_letter_df = pd.DataFrame(model_non_linear_GS.cv_results_)
GCV_letter_df
```

```python
GCV_letter_df['param_C'] = GCV_letter_df['param_C'].astype('int')
```

```python
print(model_non_linear_GS.best_params_)
print(model_non_linear_GS.best_score_)
```

```python
# plotttng for graph starts now

plt.figure(figsize=(16,8))

# plotting 1/3
plt.subplot(131)
gamma_01 = GCV_letter_df[GCV_letter_df['param_gamma'] == 0.01]

plt.plot(gamma_01['param_C'],gamma_01['mean_test_score'])
plt.plot(gamma_01['param_C'],gamma_01['mean_train_score'])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title('Gamma = 0.01')
plt.legend(['mean_test_score','mean_train_score'],loc='upper left')
plt.xscale('log')
plt.ylim([0.8,1])


# plotting 2/3
plt.subplot(132)
gamma_001 = GCV_letter_df[GCV_letter_df['param_gamma'] == 0.001]

plt.plot(gamma_001['param_C'],gamma_001['mean_test_score'])
plt.plot(gamma_001['param_C'],gamma_001['mean_train_score'])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title('Gamma = 0.001')
plt.legend(['mean_test_score','mean_train_score'],loc='upper left')
plt.xscale('log')
plt.ylim([0.8,1])


# plotting 3/3
plt.subplot(133)
gamma_0001 = GCV_letter_df[GCV_letter_df['param_gamma'] == 0.0001]
plt.plot(gamma_0001['param_C'],gamma_0001['mean_test_score'])
plt.plot(gamma_0001['param_C'],gamma_0001['mean_train_score'])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title('Gamma = 0.0001')
plt.legend(['mean_test_score','mean_train_score'],loc='upper left')
plt.xscale('log')
plt.ylim([0.8,1])
```

```python
best_nl_model = SVC(C=1000,gamma=0.01,kernel='rbf')
best_nl_model.fit(X_train,y_train)
nl_pred = best_nl_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,mean_absolute_error,mean_squared_error
```

```python
print(classification_report(y_test,linear_pred))
print(confusion_matrix(y_test,linear_pred))
print(accuracy_score(y_test,linear_pred))
```

```python
print(classification_report(y_test,nl_pred))
print(confusion_matrix(y_test,nl_pred))
print(accuracy_score(y_test,nl_pred))
```
